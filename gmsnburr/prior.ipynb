```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import halfcauchy, gamma, lognorm
```

## Half-Cauchy

### Half-Cauchy(10e5)

```python
scale = 10e5  
x = np.linspace(0, 10e5, 1000) 

# Menghitung PDF
pdf = halfcauchy.pdf(x, scale=scale)

# Visualisasi
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label=f'Half-Cauchy(10e5)', linewidth=2)
plt.title('FKP Distribusi Half-Cauchy(10e5)')
plt.xlabel('x')
plt.ylabel('Kepadatan')

plt.grid(True, alpha=0.1)
plt.legend(frameon=False)
plt.show()
```

### Half-Cauchy(3)

```python
scale = 3 
x = np.linspace(0, 10, 1000) 

# Menghitung PDF
pdf = halfcauchy.pdf(x, scale=scale)

# Visualisasi
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label=f'Half-Cauchy(3)', linewidth=2)
plt.title('FKP Distribusi Half-Cauchy(3)')
plt.xlabel('x')
plt.ylabel('Kepadatan')
plt.grid(True, alpha=0.1)
plt.legend(frameon=False)
plt.show()
```

## Gamma

### Gamma(3;0,5)

```python
scale = 0.5 
shape = 3
x = np.linspace(0, 8, 1000)  

# Menghitung PDF
pdf = gamma.pdf(x, a=shape, scale=scale)

# Visualisasi
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label=f'Gamma(3;0,5)', linewidth=2)
plt.title('FKP Distribusi Gamma(3;0,5)')
plt.xlabel('x')
plt.ylabel('Kepadatan')
plt.ylim(bottom=0)
plt.grid(True, alpha=0.1)
plt.legend(frameon=False)
plt.show()
```

### Gamma(4,2)

```python
scale = 2  
shape = 4
x = np.linspace(0, 25, 1000) 

# Menghitung PDF
pdf = gamma.pdf(x, a=shape, scale=scale)

# Visualisasi
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label=f'Gamma(4,2)', linewidth=2)
plt.title('FKP Distribusi Gamma(4,2)')
plt.xlabel('x')
plt.ylabel('Kepadatan')
plt.ylim(bottom=0)
plt.grid(True, alpha=0.1)
plt.legend(frameon=False)
plt.show()
```

## LogNormal

### LogNormal(0,75;0,75)

```python
sigma = 0.75
scale = 0.75
x = np.linspace(0, 5, 1000)

# Menghitung PDF
pdf = lognorm.pdf(x, s=sigma, scale=scale)

# Visualisasi
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label=f'LogNormal(0,75;0,75)', linewidth=2)
plt.title('FKP Distribusi LogNormal(0,75;0,75)')
plt.xlabel('x')
plt.ylabel('Kepadatan')
plt.ylim(bottom=0)
plt.grid(True, alpha=0.1)
plt.legend(frameon=False)
plt.show()
```
